```python
from src import DataModule
import matplotlib.pyplot as plt
import random
```

```python
dm = DataModule()

dm.setup()
```

```python
cols, rows = 5, 2  
fig, axs = plt.subplots(rows, cols, figsize=(20, 10))  
for col in range(cols):  
    ix = random.randint(0, len(dm.train_ds) - 1)
    image, mask = dm.train_ds[ix]
    ax = plt.subplot(rows, cols, col + 1) 
    ax.imshow(image)
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols)  
    ax.imshow(mask)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
cols, rows = 5, 4  
fig, axs = plt.subplots(rows, cols, figsize=(20, 10))  
for col in range(cols):  
    ix = random.randint(0, len(dm.train_ds) - 1)
    image, mask = dm.train_ds[ix]
    ax = plt.subplot(rows, cols, col + 1) 
    ax.imshow(image[...,:3])
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols)  
    ax.imshow(mask)
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols*2)  
    ax.imshow(image[...,3])
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols*3)  
    ax.imshow(image[...,4])
    ax.axis('off')
plt.tight_layout()
plt.show()
```
