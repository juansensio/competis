```python
from src import DataModule
import matplotlib.pyplot as plt
import random
```

```python
dm = DataModule(Dataset="DatasetFCIm")

dm.setup()
```

```python
cols, rows = 5, 4  
fig, axs = plt.subplots(rows, cols, figsize=(15, 10))  
for col in range(cols):  
    ix = random.randint(0, len(dm.train_ds) - 1)
    image, mask = dm.train_ds[ix]
    ax = plt.subplot(rows, cols, col + 1) 
    ax.imshow(image[...,:3])
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols)  
    ax.imshow(image[...,5])
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols*2)  
    ax.imshow(mask)
    ax.axis('off')
    ax = plt.subplot(rows, cols, col + 1 + cols*3)  
    inter = (mask > 0) & (image[...,5] > 0)
    ax.imshow(inter)
    ax.axis('off')
    ax.set_title(inter.sum())
plt.tight_layout()
plt.show()
```
